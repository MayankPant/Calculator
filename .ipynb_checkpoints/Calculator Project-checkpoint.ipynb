# Calculator version 1.0.1

```python
from tkinter import *

root = Tk()
root.title("Simple Calculator Version 1.0.1")
root.iconbitmap("calculator.ico")

frame_display = LabelFrame(root, padx = 7, pady = 7, bg = "#fb88f8")
frame_display.grid(row = 0, column = 0, columnspan = 2)

entry = Entry(frame_display, width = 24, bg = "#05ecf2", fg = "Red",font = ("Ink Free",30))
entry.grid(row = 0, column = 0)

frame_buttons = LabelFrame(root,padx = 28, pady = 20,bg = "#fb88f8")
frame_buttons.grid(row = 1,column = 0)

frame_del = LabelFrame(root,padx = 10, pady = 15,bg = "#fb88f8")
frame_del.grid(row = 2,column = 0)

frame_func = LabelFrame(root,padx = 15, pady = 33,bg = "#fb88f8")
frame_func.grid(row = 1,column = 1, rowspan = 5)

def button_click(number):
        current = entry.get()
        entry.delete(0,END) ## deletes whatever is on the textarea
        entry.insert(0,(current + str(number)))
        
        
    
def button_clear():
    entry.delete(0,END)
    
def button_operator(operator):
    current = entry.get()
    entry.delete(0,END)
    entry.insert(0,current + operator)




def button_del():
    current = entry.get()
    entry.delete(len(current)-1)
    

    

import math



def hcf_calc():
    # the argument is spac seperated from function name 
    expression = entry.get()
    numbers = expression.split(" ")[1:] ## 0th index has the function name
    a,b = float(numbers[0]),float(numbers[1])
    hcf = 0
    if not (a == 0) and not( b == 0):
        hcf = gcd(a,b)
        entry.delete(0,END)
        entry.insert(0,str(hcf))
    else:
        entry.delete(0,END)
        entry.insert(0,"HCF not defined for 0")
    
def gcd(a,b):
    R = a%b if a > b else b%a
    while R > 0:
        a = b
        b = R
        R = a%b if a > b else b%a
    return b if a>b else a
        
def lcm_calc():
    # the argument is spac seperated from function name 
    
    # as a*b = gcd(a,b) * lcm(a,b)
    expression = entry.get()
    numbers = expression.split(" ")[1:]
    a,b = float(numbers[0]),float(numbers[1])
    hcf = 0
    if not (a == 0) and not( b == 0):
        hcf = gcd(a,b)
        lcm = a*b/hcf
        lcm = round(lcm,2)
        entry.delete(0,END)
        entry.insert(0,str(lcm))
    else:
        entry.delete(0,END)
        entry.insert(0,"LCM not defined for 0")

def log_calc():
    # the argument is spac seperated from function name 
    expression = entry.get()
    numbers = expression.split(" ")[1:]
    
    entry.delete(0,END)
    operand = float(numbers[0])
    if not(operand <= 0):
        entry.insert(0,math.log(operand,10))
    else:
        entry.insert(0,"Your domain cannot be zero or negative")
    






def is_alphanumeric(op):
    if op == '.':
        return True
    else:
        op = ord(op)
        return 48 <= op <= 57 or 65 <= op <= 90 or 97 <= op <= 122


def is_operator(op):

    return op == '~' or op == '#' or op == '+' or op == '-' or op == '*' or op == '/' or op == '(' or op == ')' or op == '^'


def is_unary(op):
    return op == '~' or op == '#'



def get_precedence(operator):
    precedence = {"#" : 20, "~" : 20, "+" : 12, "-" : 12, "*": 13, "/" : 13, "^" : 16,"(" : 17, ")" : 17}
        ## here # is for unary plus and ~ is for unary minus
    return precedence[operator]

def perform(operator, x, y):
    x = float(x)
    y = float(y)
    m = math.modf(x) ## returns the fractional and integer part as a two item tuple
    n = math.modf(y)

    if not (m[0] == 0): ##  fractional part means a real number
        x = float(x)
    else:
        x = int(x)
    if not (n[0] == 0):
        y = float(y)
    else:
        y = int(y)

    operation = {"+" : lambda a,b : a+b, "-" : lambda a,b : a-b, "*" : lambda a,b : a*b, "/" : lambda a,b : a/b , "^" :lambda a,b : a**b}

    func = operation[operator]
    result = func(x,y)
    return result


def perform_unary(operator,x):
    x = float(x)
    m = math.modf(x)

    if not (m[0] == 0):
        x = float(x)
    else:
        x = int(x)

    unary_operation = {"#" : lambda a:+a, "~" : lambda a : -a}
    func = unary_operation[operator]
    result = func(x)
    return result

def balanced(infix):
    stack = []
    
    for op in infix:
        if op == '(':
            stack.append(op)
        elif op == ')':
            stack.pop()
    
    return len(stack) == 0



def button_equal():
    try:
        infix = entry.get()

        # checking for functional inputs
        if "hcf" in infix:
            hcf_calc()
            return
        if "lcm" in infix:
            lcm_calc()
            return
        if "log" in infix:
            log_calc()
            return

        ## checking for balanced infix expression

        if not balanced(infix):
            entry.delete(0,END)
            entry.insert(0,"Please enter a valid balanced infix expression without spaces")
            return

        operand = []
        operator = []
        count = ""
        unary = True
        for op in infix:

            if is_operator(op):
                if len(count) > 0:
                    operand.append(count)
                    count = ""
                    unary = False ## once you have an operator , after than only a binary operator can come
                if op == '(':
                    operator.append(op)
                elif op == ')':
                    while not (len(operator) == 0) and not operator[len(operator)-1] == '(':
                        a = operator.pop()
                        if is_unary(a):
                            x = operand.pop()
                            result = perform_unary(a, x)
                            operand.append(result)
                        else:
                            x = operand.pop()
                            y = operand.pop()
                            result = perform(a,y,x)
                            operand.append(result)
                    operator.pop()

                else:
                    if unary:
                        if op == '-':
                            op = '~'
                        elif op == '+':
                            op = '#'
                    while not (len(operator) == 0) and get_precedence(operator[len(operator)-1]) >= get_precedence(op) and not operator[len(operator)-1] == '(':

                            a = operator.pop()
                            if is_unary(a):
                                x = operand.pop()
                                result = perform_unary(a, x)
                                operand.append(result)

                            else:
                                x = operand.pop()
                                y = operand.pop()

                                result = perform(a, y, x)
                                operand.append(result)
                    operator.append(op)
                    unary = True  ## once the binary operator is pushed , now you can have a unary operator or an operand


            elif is_alphanumeric(op):
                count = count + op



        if len(count) > 0:
            operand.append(count)


        while not (len(operator) == 0):

            a = operator.pop()
            if is_unary(a):
                x = operand.pop()
                result = perform_unary(a, x)
                operand.append(result)

            else:
                x = operand.pop()
                y = operand.pop()

                result = perform(a, y, x)
                operand.append(result)


        entry.delete(0,END)
        entry.insert(0,str(operand.pop()))
        
    except:
        entry.delete(0,END)
        entry.insert(0,"An Error Occured")
        

    
    
    
    
def button_func(function):
    entry.delete(0,END)
    entry.insert(0,function)
    
    
    
## creating number buttons
button_1 = Button(frame_buttons, text = "1",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(1))
button_2 = Button(frame_buttons, text = "2",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(2))
button_3 = Button(frame_buttons, text = "3",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(3))
button_4 = Button(frame_buttons, text = "4",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(4))
button_5 = Button(frame_buttons, text = "5",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(5))
button_6 = Button(frame_buttons, text = "6",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(6))
button_7 = Button(frame_buttons, text = "7",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(7))
button_8 = Button(frame_buttons, text = "8",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(8))
button_9 = Button(frame_buttons, text = "9",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(9))
button_0 = Button(frame_buttons, text = "0",padx = 40, pady = 20, bg = "#05ecf2", command = lambda : button_click(0))
button_decimal = Button(frame_buttons, text = ".",padx = 41, pady = 20, bg = "#05ecf2", command = lambda : button_click("."))
button_equal = Button(frame_buttons, text = "=",padx =40, pady = 20, bg = "#05ecf2", command = button_equal)

## creating clearing buttons

button_clear = Button(frame_del, text = "CLEAR",padx = 30, pady = 20, bg = "#200e71" ,fg = "#ffffff", command = button_clear)
button_del = Button(frame_del, text = "DEL",padx = 40, pady = 20, bg = "#200e71" ,fg = "#ffffff", command = button_del)
button_space = Button(frame_del, text = "SPC",padx = 40, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_operator(" "))
## creating functional buttons


button_left_parenthesis = Button(frame_func, text = "(",padx = 41, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_click('('))
button_right_parenthesis = Button(frame_func, text = ")",padx = 41, pady = 21, bg = "#200e71",fg = "#ffffff" , command = lambda : button_click(')'))
button_add = Button(frame_func, text = "+",padx = 40, pady = 21, bg = "#200e71",fg = "#ffffff",  command = lambda : button_operator("+"))
button_sub = Button(frame_func, text = "-",padx = 42, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_operator("-"))
button_mul = Button(frame_func, text = "*",padx = 42, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_operator("*"))
button_div = Button(frame_func, text = "/",padx = 42, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_operator("/"))
button_power = Button(frame_func, text = "^",padx = 40, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_operator("^"))


## functional keys
button_log = Button(frame_func, text = "log",padx = 35, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_func("log "))
button_hcf = Button(frame_func, text = "hcf",padx = 35, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_func("hcf "))
button_lcm = Button(frame_func, text = "lcm",padx = 34, pady = 21, bg = "#200e71" ,fg = "#ffffff", command = lambda : button_func("lcm "))


## placing number buttons inside frame_buttons

# 789
button_7.grid(row = 1, column = 0)
button_8.grid(row = 1, column = 1)
button_9.grid(row = 1, column = 2)

# 456
button_4.grid(row = 2, column = 0)
button_5.grid(row = 2, column = 1)
button_6.grid(row = 2, column = 2)

# 123
button_1.grid(row = 3, column = 0)
button_2.grid(row = 3, column = 1)
button_3.grid(row = 3, column = 2)

# 0 . %
button_0.grid(row = 4, column = 1)
button_decimal.grid(row = 4, column = 0)
button_equal.grid(row = 4,column = 2)

# placing clearing buttons

button_clear.grid(row = 5, column = 0)
button_del.grid(row = 5, column = 2)
button_space.grid(row = 5, column = 1)
#placing functional buttons inside frame_func

button_add.grid(row = 0, column = 0)
button_sub.grid(row = 1, column = 0)
button_mul.grid(row = 2, column = 0)
button_div.grid(row = 3, column = 0)
button_power.grid(row = 4, column = 0)

button_left_parenthesis.grid(row = 0, column = 1)
button_right_parenthesis.grid(row = 1, column = 1)
button_hcf.grid(row = 2, column = 1)
button_lcm.grid(row = 3, column = 1)
button_log.grid(row = 4, column = 1)


root.mainloop()
```
